````python
import sqlite3

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
````

````python
cursor.execute("""
CREATE TABLE Student(
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Age INTEGER
)
""")

cursor.executemany(
    "INSERT INTO Student VALUES (?, ?, ?)",
    [
        (1, 'John', 20),
        (2, 'Alice', 21),
        (3, 'Bob', 22)
    ]
)

conn.commit()
print("Database created successfully")
````
